# Slide 1: Introdução às Leis de De Morgan

As **Leis de De Morgan** são um par de regras de transformação que são regras válidas de inferência. Elas recebem o nome de Augustus De Morgan, um matemático britânico do século XIX.

Essas leis relacionam os três operadores básicos da lógica e teoria dos conjuntos:
1.  **Negação** (Complemento no contexto de conjuntos).
2.  **Conjunção** (Interseção).
3.  **Disjunção** (União).

Nesta apresentação, vamos explorar como essas leis funcionam e prová-las usando a Teoria dos Conjuntos.

# Slide 2: As Duas Leis de De Morgan

Em termos de conjuntos, para quaisquer dois conjuntos $A$ e $B$ dentro de um conjunto universo $U$, as leis são expressas como:

1.  **O complemento da união é a interseção dos complementos:**
    $$(A \cup B)^c = A^c \cap B^c$$

2.  **O complemento da interseção é a união dos complementos:**
    $$(A \cap B)^c = A^c \cup B^c$$

Onde $X^c$ denota o complemento de $X$ (também escrito como $\bar{X}$ ou $U \setminus X$).

# Slide 3: Intuição Visual e Lógica

A ideia central é que a negação de uma disjunção (OU) é equivalente à conjunção (E) das negações.

*   **Exemplo Lógico:** Dizer que "não é verdade que hoje é feriado OU amanhã é domingo" é o mesmo que dizer que "hoje NÃO é feriado E amanhã NÃO é domingo".
*   **Exemplo de Conjunto:** Estar fora da área ocupada por $A$ ou $B$ significa estar simultaneamente fora de $A$ e fora de $B$.

# Slide 4: Prova da Primeira Lei (Parte 1)

**Teorema:** $(A \cup B)^c = A^c \cap B^c$

Para provar que dois conjuntos são iguais, devemos mostrar que cada um é subconjunto do outro:
1.  $(A \cup B)^c \subseteq A^c \cap B^c$
2.  $A^c \cap B^c \subseteq (A \cup B)^c$

Vamos começar com a primeira inclusão. Seja um elemento $x \in (A \cup B)^c$.

# Slide 5: Prova da Primeira Lei (Parte 2)

**Passo 1: Mostrar $(A \cup B)^c \subseteq A^c \cap B^c$**

*   Se $x \in (A \cup B)^c$, então $x \notin (A \cup B)$ por definição de complemento.
*   Se $x \notin (A \cup B)$, então $x$ não pode estar em $A$ nem em $B$. Ou seja: $x \notin A$ **e** $x \notin B$.
*   Pela definição de complemento: $x \in A^c$ **e** $x \in B^c$.
*   Portanto, $x \in A^c \cap B^c$.

Isso prova que $(A \cup B)^c \subseteq A^c \cap B^c$.

# Slide 6: Prova da Primeira Lei (Parte 3)

**Passo 2: Mostrar $A^c \cap B^c \subseteq (A \cup B)^c$**

*   Seja $y \in A^c \cap B^c$.
*   Então $y \in A^c$ **e** $y \in B^c$.
*   Isso implica que $y \notin A$ **e** $y \notin B$.
*   Se $y$ não está em nenhum dos dois, então $y$ não pode estar na união deles: $y \notin (A \cup B)$.
*   Logo, $y \in (A \cup B)^c$.

Como provamos as duas inclusões, concluímos que $(A \cup B)^c = A^c \cap B^c$.

# Slide 7: Prova da Segunda Lei (Parte 1)

**Teorema:** $(A \cap B)^c = A^c \cup B^c$

Agora vamos provar a relação para a interseção. O procedimento é análogo, mostrando que:
1.  $(A \cap B)^c \subseteq A^c \cup B^c$
2.  $A^c \cup B^c \subseteq (A \cap B)^c$

# Slide 8: Prova da Segunda Lei (Parte 2)

**Passo 1: Mostrar $(A \cap B)^c \subseteq A^c \cup B^c$**

*   Seja $x \in (A \cap B)^c$. Então $x \notin (A \cap B)$.
*   Isso significa que é falso que $x$ está em $A$ e em $B$ simultaneamente.
*   Portanto, ou $x$ não está em $A$, ou $x$ não está em $B$ (ou em nenhum dos dois).
*   $x \notin A \lor x \notin B \implies x \in A^c \lor x \in B^c$.
*   Logo, $x \in A^c \cup B^c$.

# Slide 9: Prova da Segunda Lei (Parte 3)

**Passo 2: Mostrar $A^c \cup B^c \subseteq (A \cap B)^c$**

*   Seja $y \in A^c \cup B^c$. Então $y \in A^c$ ou $y \in B^c$.
*   Isso implica que $y \notin A$ ou $y \notin B$.
*   Se $y$ falta em pelo menos um deles, ele não pode estar na interseção dos dois: $y \notin (A \cap B)$.
*   Logo, $y \in (A \cap B)^c$.

As duas inclusões confirmam que $(A \cap B)^c = A^c \cup B^c$.

# Slide 10: Conclusão e Aplicações

As Leis de De Morgan são fundamentais em:
*   **Circuitos Digitais:** Simplificação de portas lógicas (Transformar portas NAND em NOR e vice-versa).
*   **Programação:** Simplificação de condicionais complexas `if(!(A && B))` torna-se `if(!A || !B)`.
*   **Matemática Pura:** Base para álgebra booleana e teoria das probabilidades.

Compreender essas leis permite transitar entre o "E" e o "OU" de forma rigorosa e eficiente.

### Exemplo Prático em Python

Vamos ver como a lei `not (A or B) == (not A and not B)` se aplica na prática para validar se um usuário preencheu campos obrigatórios.

```python
def validar_entrada(nome, email):
    # 1ª Lei: !(A ou B) é o mesmo que (!A e !B)

    # Expressão Original: 'Se não for verdade que (tem nome ou tem email)'
    if not (nome or email):
        print("Erro: Ambos os campos estão vazios!")
    else:
        print("Sucesso: Pelo menos um campo preenchido.")

    # Expressão Aplicando De Morgan: 'Se não tem nome E não tem email'
    if not nome and not email:
        print("[De Morgan] Erro: Ambos os campos estão vazios!")

# Teste
print("--- Testando com valores vazios ---")
validar_entrada("", "")

print("\n--- Testando com valores preenchidos ---")
validar_entrada("Alice", "")
```

### Exemplo da Segunda Lei de De Morgan

A lei `not (A and B) == (not A or not B)` é útil para verificar falhas. Se para um sistema funcionar é necessário ter **Internet E Bateria**, o sistema falha se **NÃO tiver Internet OU NÃO tiver Bateria**.

```python
def verificar_sistema(tem_internet, tem_bateria):
    # 2ª Lei: !(A e B) é o mesmo que (!A ou !B)

    # Lógica de Sucesso: O sistema só funciona se ambos forem True
    sistema_online = tem_internet and tem_bateria

    # Expressão Original: 'Se NÃO (tem internet E tem bateria)'
    if not (tem_internet and tem_bateria):
        print("Alerta: Sistema offline por falta de requisitos!")

    # Expressão Aplicando De Morgan: 'Se NÃO tem internet OU NÃO tem bateria'
    if not tem_internet or not tem_bateria:
        print("[De Morgan] Alerta: Falha detectada em pelo menos um componente!")

# Teste: Faltando apenas um requisito
print("--- Testando com Internet, mas sem Bateria ---")
verificar_sistema(True, False)
```
